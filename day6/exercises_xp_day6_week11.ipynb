```python
# Exercise 2
import numpy as np
# Sample dataset
data = [12, 15, 13, 12, 18, 20, 22, 21]

# Calculate mean
mean_value = np.mean(data)
print("Mean:", mean_value)

# Calculate median
median_value = np.median(data)
print("Median:", median_value)

# Calculate variance
variance_value = np.var(data, ddof=1)  # ddof=1 for sample variance
print("Variance:", variance_value)

# Calculate standard deviation
std_deviation = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
print("Standard Deviation:", std_deviation)
```

```python
# Exercise 3
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mean = 50
std_dev = 10

# Generate a range of values
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Generate the normal distribution
pdf = norm.pdf(x, mean, std_dev)

# Plot the normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Normal Distribution\nMean = {mean}, Std Dev = {std_dev}')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Exercise 4
# Parameters for the normal distributions
from scipy import stats

mean1, std_dev1, size1 = 50, 10, 100
mean2, std_dev2, size2 = 55, 10, 100

# Generate two sets of random data
data1 = np.random.normal(loc=mean1, scale=std_dev1, size=size1)
data2 = np.random.normal(loc=mean2, scale=std_dev2, size=size2)

# Perform a T-test
t_stat, p_value = stats.ttest_ind(data1, data2, equal_var=False)  # Use equal_var=False for Welch's T-test

# Print the results
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
# Exercise 6
# Generate a random dataset from a normal distribution
data = np.random.normal(0, 1, 1000)

# Calculate kurtosis
kurtosis_value = stats.kurtosis(data)
print(f"Kurtosis: {kurtosis_value}")

# Calculate skewness
skewness_value = stats.skew(data)
print(f"Skewness: {skewness_value}")

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Add a title and labels
plt.title('Histogram of Randomly Generated Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
