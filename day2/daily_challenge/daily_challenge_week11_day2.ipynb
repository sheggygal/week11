```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)  # for reproducibility
temperature_data = np.random.uniform(-5, 35, (10, 12))
```

```python
cities = [f'City_{i+1}' for i in range(10)]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
df = pd.DataFrame(temperature_data, index=cities, columns=months)
df
```

```python
# Calculate annual average temperature for each city
annual_avg_temp = df.mean(axis=1)
for city in cities:
    print(f"Annual average temperature for {city}: {annual_avg_temp[city]}")
```

```python
highest_avg_temp_city = annual_avg_temp.idxmax()
lowest_avg_temp_city = annual_avg_temp.idxmin()
print(f"City with highest average temperature: {highest_avg_temp_city}")
print(f"City with lowest average temperature: {lowest_avg_temp_city}")
```

```python
plt.figure(figsize=(12, 8))

for city in df.index:
    plt.plot(df.columns, df.loc[city], label=city)

plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.title('Monthly Temperature Trends for Each City')
plt.legend()
plt.show()
```
